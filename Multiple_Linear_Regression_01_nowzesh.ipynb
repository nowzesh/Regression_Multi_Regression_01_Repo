```python
# Import python libraries (Codes by A Z M Nowzesh Hasan)
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
# Uploading dataset
dataset = pd.read_csv('50_Startups.csv')

# View top 5 rows
dataset.head()
```

```python
# All columns 
dataset.columns
```

```python
# Checking columnwise null values
dataset.isnull().sum()
```

```python
# Feature extraction 
x1 = dataset.iloc[:, :-1].values
y = dataset.iloc[:, -1:].values
print(f"x: {x1[:10]}\n\ny: {y[:10]}") # Showing just top 10 rows 
```

```python
# Move last column to first column 

x = np.concatenate((x1[:, -1][:, np.newaxis], x1[:, :-1]), axis=1)

print(x[:10])
```

```python
# Encoding independent variables: One-Hot_Encoding: 4th column: "state" 
# To convert categorical (non-numeric) data into a numeric format

# Import python libraries 
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer 

ct = ColumnTransformer(transformers= [('encoder', OneHotEncoder(), [0])], remainder = 'passthrough') # remainder='passthrough': All other columns are left unchanged
x = np.array(ct.fit_transform(x))
print(f"Feature encoded (OneHotEncoder): \n{x[:10]}")

# To convert categorical (non-numeric) data into a numeric format
```

```python
# Split data: Train & Test Sets 

# Import Python Libraries 
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x,y, test_size = 1/3, random_state = 0) 
# random_state = 0 means it will split data similar way for every run, no randomness, split reproducible & consistent 
print(f"x-train: {x_train[:10]}\n\ny-train: {y_train[:10]}\n\nx-test: {x_test[:10]}\n\ny-test: {y_test[:10]}") # Showing just top 10 rows 
```

```python
# Linear Regression Model 

# Importing Pythin Library 
from sklearn.linear_model import LinearRegression

# Training dataset into Linear Regression Model 
regressor = LinearRegression()
regressor.fit(x_train, y_train)
```

```python
# Prediction values from Linear Regression Model After Training 
y_pred = regressor.predict(x_test)
plt.scatter(y_test, y_pred)
plt.xlabel('Actual Profit')
plt.ylabel('Predicted Profit')
plt.title('Actual vs Predicted Profit')

# Annotate each point with actual and predicted values
for actual, predicted in zip(y_test, y_pred):
    plt.text(actual, predicted, f'({int(actual[0])}, {int(predicted)})', fontsize=8)

plt.plot([min(y_test), max(y_test)], [min(y_test), max(y_test)], color='red', linestyle='--')  # reference line
plt.grid(True)
plt.tight_layout()
plt.show()

```

```python
# Ensure arrays are column vectors
y_test_array = np.array(y_test).reshape(-1, 1)
y_pred_array = np.array(y_pred).reshape(-1, 1)

# Combine side by side
combined = np.hstack((y_test_array, y_pred_array))

# Print header
print(f"{'Actual':>15} {'Predicted':>15}")

# Print each row
for actual, predicted in combined:
    print(f"{actual:15.2f} {predicted:15.2f}")
```

```python
# Single prediction from user input values 

# Input from user
rd_spend = float(input("Enter R&D Spend: "))
admin = float(input("Enter Administration Spend: "))
marketing = float(input("Enter Marketing Spend: "))
state = input("Enter State (e.g., California, Florida, New York): ")

# New data in correct format
new_data = np.array([[state, rd_spend, admin, marketing]])
print(new_data)

# Transform with trained ColumnTransformer

new_data_transformed = ct.transform(new_data)

print(new_data_transformed)


# Make sure data type is numeric (not object or string)
new_data_transformed = new_data_transformed.astype(float)

# Predict
predicted_profit = regressor.predict(new_data_transformed)
print(f"\nPredicted Profit: ${predicted_profit[0][0]:,.2f}")
```

```python
# Getting the final linear regression equation with the values of the coefficients
print(regressor.coef_)
print(regressor.intercept_)
```

```python
# Evaluation of the model performance 

from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import numpy as np

mae = mean_absolute_error(y_test, y_pred) # Mean Absolute Error (MAE): Average of absolute errors (|actual - predicted|)
mse = mean_squared_error(y_test, y_pred) # Mean Squared Error (MSE): Penalizes larger errors more than MAE
rmse = np.sqrt(mse) # Root Mean Squared Error (RMSE): RMSE is in same units as the output.
r2 = r2_score(y_test, y_pred) # R2: 0 to 1 (higher is better) 

print(f"MAE: {mae}")
print(f"MSE: {mse}")
print(f"RMSE: {rmse}")
print(f"R²: {r2}") 
```
